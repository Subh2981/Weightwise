```python
x=df.drop('ObesityCategory',axis=1)
y=df['ObesityCategory']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
le=LabelEncoder()
```

```python
x_train['Gender']=le.fit_transform(x_train['Gender']).astype(int)
```

```python
x_train.head()
```

```python
df.iloc[82]
```

```python
x_test['Gender']=le.transform(x_test['Gender']).astype(int)
```

```python
x_test.head()
```

```python
y_train=le.fit_transform(y_train.values).astype(int)
```

```python
y_test=le.fit_transform(y_test.values).astype(int)
```

```python
Scaler=StandardScaler()
```

```python
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
x_train
```

```python
LR=LogisticRegression()
```

```python
model=LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
y_pred_tr=LR.predict(x_train)
```

```python
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_train,y_pred_tr)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_train,y_pred_tr))
```

```python
param_grid={
    'n_neighbors':[3,5,7,9,13,15],
    'weights':['uniform','distance'],
    'n_jobs':[1]   
}
knn=KNeighborsClassifier()
cv=StratifiedKFold(n_splits=5)
grid=GridSearchCV(knn,param_grid=param_grid,cv=cv,scoring='f1_macro')
```

```python
grid.fit(x_train,y_train)
```

```python
df['ObesityCategory'].value_counts()
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
