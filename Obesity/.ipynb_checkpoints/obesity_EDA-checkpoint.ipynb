```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("obesity_data.csv")
```

```python
avrage_ob_age=df.groupby(['ObesityCategory','Gender'])['Age'].mean()
```

```python
data=avrage_ob_age.reset_index()
```

```python
sns.barplot(data=data,x='ObesityCategory',y='Age',hue='Gender')
```

```python
data
```

```python
avg_bmi=df.groupby(['ObesityCategory','Gender'])['BMI'].mean()
```

```python
df.columns
```

```python
avg_bmi=avg_bmi.reset_index()
```

```python
sns.barplot(data=avg_bmi,x='ObesityCategory',y='BMI',hue='Gender')
```

```python
avg_bmi
```

```python
df.PhysicalActivityLevel.unique()
```

```python
df.columns
```

```python
avg_bmi_level = (
    df.groupby(['ObesityCategory','Gender'])['PhysicalActivityLevel']
      .apply(lambda x: x.mode().iloc[0])
      .reset_index()
)
avg_bmi_level=avg_bmi_level.reset_index()
```

```python
sns.barplot(data=avg_bmi_level,x='ObesityCategory',y='PhysicalActivityLevel',hue='Gender',palette='rainbow')
```
